```python
import pandas as pd
```

```python
train_data = pd.read_csv("/content/drive/MyDrive/Datasets-20240614T041107Z-001/Datasets/titanic/train.csv")
test_data = pd.read_csv("/content/drive/MyDrive/Datasets-20240614T041107Z-001/Datasets/titanic/test.csv")
```

```python
train_data.head()
```

```python
print(train_data.columns.values)
```

```python
train_data = train_data[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch','Fare']]
train_data_X = train_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch','Fare']]
train_data_y = train_data[['Survived']]
```

```python
train_data_X.tail()
```

```python
train_data_y.head()
```

```python
train_data_X.loc[:,'Sex'] = train_data_X['Sex'].replace({'male': 0, 'female': 1})
```

```python
print(train_data_X.isnull().sum())
```

```python
print(train_data_y.isnull().sum())
```

```python
train_data_X.loc[:,'Age'] = train_data_X['Age'].fillna(train_data_X['Age'].mean())
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
logreg = LogisticRegression()
logreg.fit(train_data_X, np.ravel(train_data_y))
```

```python
test_data.head()
```

```python
test_data = test_data[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]]
test_data.loc[:,'Sex'] = test_data['Sex'].replace({'male': 0, 'female': 1})
test_data.loc[:,'Age'] = test_data['Age'].fillna(test_data['Age'].mean())
test_data.loc[:,'Fare'] = test_data['Fare'].fillna(test_data['Fare'].mean())
```

```python
y_pred = logreg.predict(test_data)
```

```python
from google.colab import files
PassengerId = list(range(892, 1310))
df = pd.DataFrame({'PassengerId': PassengerId, 'Survived': y_pred})
df.to_csv('submission-six.csv', index=False)
files.download('submission-six.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_data_X, train_data_y, test_size=0.2, random_state=42)
```

```python
logreg.fit(X_train, y_train)
```

```python
y_val_pred = logreg.predict(X_val)
```

```python
accuracy = accuracy_score(y_val, y_val_pred)
precision = precision_score(y_val, y_val_pred)
recall = recall_score(y_val, y_val_pred)
f1 = f1_score(y_val, y_val_pred)

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')

# Confusion Matrix
conf_matrix = confusion_matrix(y_val, y_val_pred)
print(f'Confusion Matrix:\n{conf_matrix}')

# Classification Report
class_report = classification_report(y_val, y_val_pred)
print(f'Classification Report:\n{class_report}')
```
